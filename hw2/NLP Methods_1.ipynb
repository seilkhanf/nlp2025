### Noisy sound

```python
# Importing the speech_recognition library
import speech_recognition as sr
```

```python
recognizer = sr.Recognizer()
```

```python
# Convert audio to AudioFile
noisy_support_call = sr.AudioFile("staytuned.wav")
```

```python
# Record the audio from the noisy support call
with noisy_support_call as source:
 # Adjust the recognizer energy threshold for ambient noise
    recognizer.adjust_for_ambient_noise(source, duration=0.5)
    noisy_support_call_audio = recognizer.record(noisy_support_call)
 
```

```python
#Transcribe the speech from the noisy support call
text = recognizer.recognize_google(noisy_support_call_audio,
                                   language="en-US")
```

# Text preprocessing

```python
punctuation='~!@#$%^&*()-_+=:;\'\",./?'
de_punct_sent=[]
for s in sentences:
    for p in punctuation:
        s=s.replace(p,' ')
    s=s.strip()
    de_punct_sent.append(s)
de_punct_sent

```

# Concat broken sentences

```python
'a'.islower()
```

```python
unbroken=[]
for i,s in enumerate(de_punct_sent):
    if s[0].islower():
        unbroken[-1]=unbroken[-1]+' '+s
    else:
        unbroken.append(s)
        
unbroken
```

# Lemmatization

```python
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer 
```

```python
nltk.download('wordnet')
```

```python
def get_pos_tag(t):
    try:
        tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV,
                    "P": wordnet.NOUN,#
                    "C": wordnet.NOUN,
                    "D": wordnet.NOUN,
                    "E": wordnet.NOUN,
                    "I": wordnet.NOUN,
                    "L": wordnet.NOUN,
                    "M": wordnet.NOUN,
                    "T": wordnet.NOUN,
                    "U": wordnet.NOUN,
                    "W": wordnet.NOUN,
                    "F": wordnet.NOUN,
                    "S": wordnet.NOUN}
        return tag_dict[t[0]]
    except:
        return wordnet.NOUN
    
```

```python
os.listdir()
```

```python
lemmatizer = WordNetLemmatizer()
print(type(lemmatizer))
```

```python
nltk.download('averaged_perceptron_tagger')
lemmatizer = WordNetLemmatizer()
lemmed_sents=[''.join([lemmatizer.lemmatize(tg[0], pos=get_pos_tag(tg[1])) for tg in nltk.pos_tag(s.split())]) for s in de_punct_sent]


print(de_punct_sent[:5])
print()
print(lemmed_sents[:5])
```

# Dimensionality reduction

```python
from sklearn.decomposition import PCA
```

```python
def get_2d_3d_rep(X):
    #2D
    pca = PCA(n_components=2)
    X_2d=pca.fit_transform(X)
    
    #3D
    pca = PCA(n_components=3)
    X_3d=pca.fit_transform(X)
    
    return X_2d, X_3d
```

```python
X_2d, X_3d=get_2d_3d_rep(X.toarray())
```

```python
X2_2d, X2_3d=get_2d_3d_rep(X2.toarray())
```
